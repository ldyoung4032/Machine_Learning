```python
import pandas as pd
import pdpipe as pdp

df = pd.read_csv('USA_Housing.csv')
```

```python
df.describe()
```

```python
df.head()
```

```python
## 실수형인 Avg. Area Number of Rooms를 서열 범주형값을 갖는 컬럼 House_size로 만든다
## small : ~25% (~6.3)
## Medium : 25%~75% (6.3 ~ 7.7)
## Big : 75%~ (7.7~)
```

```python
def room_size(n) :
    if n <= 6.3 :
        return 'Small'
    elif 6.3 < n <= 7.7 :
        return 'Medium'
    else :
        return 'Big'
    
df['House_size'] = df['Avg. Area Number of Rooms'].apply(room_size)
```

```python
df.head()
```

```python
## 컬럼 삭제 파이프라인 객체 만들기
drop_age = pdp.ColDrop('Avg. Area Number of Rooms')
df2 = drop_age(df)
```

```python
df2.head()
```

```python
## 파이프라인 추가
```

```python
pipeline = pdp.ColDrop('Avg. Area Number of Rooms')
pipeline = pdp.OneHotEncode('House_size')

df3 = pipeline(df)
```

```python
df3.head()
```

```python
## 로우 삭제 파이프라인 객체 만들기
## Price컬럼의 값이 250000미안인 로우 삭제
```

```python
def price_tag(x) :
    if x > 250000 :
        return 'keep'
    else :
        return 'drop'
```

```python
pipeline += pdp.ApplyByCols('Price', price_tag, 'Price_tag', drop = False)
df4 = pipeline(df)
print('df 로우 수 : ', len(df), '   df4 로우 수 : ', len(df4), sep = '')
df4.head()
```

```python
pipeline += pdp.ValDrop(['drop'], 'Price_tag')
df5 = pipeline(df)
print('df 로우 수 : ', len(df), '   df5 로우 수 : ', len(df5), sep = '')
df5.head()
```

```python
## 임시컬럼 Price_tag 삭제

pipeline += pdp.ColDrop('Price_tag')
df6 = pipeline(df)
print('df 로우 수 :', len(df), '   df6 로우 수 : ', len(df6), sep = '')
df6.head()
```
