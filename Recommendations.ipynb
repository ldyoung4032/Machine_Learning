# 콘텐츠 기반 필터링

```python
import pandas as pd
import numpy as np
import warnings; warnings.filterwarnings('ignore')

movies = pd.read_csv('./tmdb_5000_movies.csv')
print(movies.shape)
movies.head()
```

```python
movies_df = movies[['id', 'title', 'genres', 'vote_average', 'vote_count', 'popularity', 'keywords', 'overview']]
movies_df
```

```python
pd.set_option('max_colwidth', 100) ## 컬럼 내용 다 볼 수 있게 늘린거
movies_df[['genres', 'keywords']][:1]
```

```python
from ast import literal_eval
movies_df['genres'] = movies_df['genres'].apply(literal_eval)
movies_df['keywords'] = movies_df['keywords'].apply(literal_eval)
```

```python
movies_df['genres'] = movies_df['genres'].apply(lambda x : [y['name'] for y in x]) ## dict들의 name키에 해당하는 데이터들이 리스티에 들어간다
movies_df['keywords'] = movies_df['keywords'].apply(lambda x : [y['name'] for y in x])
movies_df[['genres', 'keywords']][:2]
```

```python
## 위의 genres를 문자열로 변환 후 칼럼을 Count기반으로 피처 벡터화 ㄱㄱ
## 이 후 변환된 ㄷ데이터 세트를 코사인 유사도를 통해 비교
## 이를 위해 데이터 세트의 레코드별로 타 레코드와 장르에서 코사인 유사도 값을 가지는 객체 생성
## 장르 유사도가 높은 영화 중에서 평점이 높은 순으로 영화를 추천합니다
```

## 장르 콘텐츠 유사도 측정

```python
from sklearn.feature_extraction.text import CountVectorizer

# CountVectorizer를 적용하기 위해 공백문자로 word 단위가 구분되는 문자열로 변환
movies_df['genres_literal'] = movies_df['genres'].apply(lambda x : (' ').join(x))
print(movies_df['genres'])
print("---------")
print(movies_df['genres_literal'])
count_vect = CountVectorizer(min_df=0, ngram_range=(1,2)) ## 0이하 추출 X , 토큰화된 단어를 1개씩, 그리고 순서대로 2개씩 묶어서 피처로 추출 (min, max)
genre_mat = count_vect.fit_transform(movies_df['genres_literal'])

print(genre_mat.shape) ## 4803 레코드 수(영화 하나하나), 276 개별 단어 피처 개수
# print(genre_mat)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim.shape)
print(genre_sim[:2])
```

```python
genre_sim_sorted_ind = genre_sim.argsort()[:,::-1] ## argsost 작은 값부터 순서대로 데이터의 인덱스값을 반환, 처음부터 끝까지 -1만큼씩(역순으로)
print(genre_sim_sorted_ind[:1]) ## 0은 자신, 자신과 가장 높은 유사도를 가진 레코드의 위치 인덱스는 3494, 가장 낮은 유사도는 2401
```

### 장르 콘텐츠 필터링을 이용한 영화 추천

```python
def find_sim_movie(df, sorted_ind, title_name, top_n = 10) :
    title_movie = df[df['title']==title_name] ## title 추출
    
    title_index = title_movie.index.values ## title_name의 데이터프레임 인덱스 객체를 ndarray로 반환
    similar_indexes = sorted_ind[title_index, :(top_n)] ## 유사도순으로 인덱스 추출
    
    print(similar_indexes)
    print("------")
    similar_indexes = similar_indexes.reshape(-1)
    print(similar_indexes)
    
    return df.iloc[similar_indexes]
    
similar_movies = find_sim_movie(movies_df, genre_sim_sorted_ind, 'The Godfather', 10)
similar_movies[['title', 'vote_average']] ## 제목, 평균 평점
```

## 가중평점 = (v/(v+m))*R + (m/(v+m))*C
### v : 개별 영화에 평점을 투표한 횟수, vote_count
### m : 평점을 부여하기 위한 최소 투표 횟수
### R : 개별 영화에 대한 평균 평점, vote_average
### C : 전체 영화에 대한 평균 평점, movie_df['vote_average'].mean()

```python
C = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(0.6) ## m값을 높이면 평점 투표 횟수가 많은 영화에 더 많은 가중 평점을 부여,  상위 60%에 해당하는 횟수를 기준으로 정함
print('C : ', round(C,3), '   m : ', round(m,3))
```

```python
percentile = 0.6
m = movies['vote_count'].quantile(percentile)
C = movies['vote_average'].mean()

def weighted_vote_average(record) :
    v = record['vote_count']
    R = record['vote_average']
    
    return ((v/(v+m))*R) + ((m/(m+v))*C)

movies_df['weighted_vote'] = movies.apply(weighted_vote_average, axis = 1)

movies_df[['title', 'vote_average', 'weighted_vote', 'vote_count']].sort_values('weighted_vote', ascending = False)[:10]
```

```python
## 위의 새로운 기준으로 영화 추천
def find_sim_movie(df, sorted_ind, title_name, top_n = 10 ) :
    title_movie = df[df['title']==title_name] ## title 추출
    title_index = title_movie.index.values
    
    similar_indexes = sorted_ind[title_index, :(top_n*2)] ## 유사도순으로 인덱스 추출
    similar_indexes = similar_indexes.reshape(-1)
    
    similar_indexes = similar_indexes[similar_indexes != title_index]
    
    return df.iloc[similar_indexes].sort_values('weighted_vote', ascending = False)[:top_n]

similar_movies = find_sim_movie(movies_df, genre_sim_sorted_ind, 'Gravity') 
similar_movies[['title', 'vote_average', 'weighted_vote']]

## Spirited Away, Inception, Interstellar, Gravity
```
